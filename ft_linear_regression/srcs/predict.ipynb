```python
import json
import os
import numpy as np
```

```python
theta0: float = 0.0
theta1: float = 0.0

if os.path.exists("model.json"):
        with open("model.json", "r") as f:
                model = json.load(f)
                theta0 = float(model["theta0"])
                theta1 = float(model["theta1"])

print(f"theta0: {theta0}")
print(f"theta1: {theta1}")
```

```python
try:
        mileage: float = float(input("Enter mileage: "))
        price = theta0 + theta1 * mileage
        print(f"estimated price for the mileage {mileage}: {price}")

except ValueError:
        print("Invalid input value")
        exit(1)
```
